```python
# -- coding: utf-8 --
"""
Created on Sun May 11 01:53:57 2025

@author: HARSHIT NARAIN
"""

import os
import numpy as np
import matplotlib.pyplot as plt
import time
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, AveragePooling2D
from tensorflow.keras.initializers import HeNormal
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Input, Concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import kagglehub

# Download Animals10 dataset
dataset_path = kagglehub.dataset_download("alessiocorrado99/animals10")
data_dir = os.path.join(dataset_path, "raw-img")
print("Dataset downloaded to:", data_dir)

# Configuration
img_height, img_width = 224, 224
batch_size = 32
epochs = 10

# Data Generators
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=True)

val_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=False)

num_classes = train_generator.num_classes

# Plot training history
def plot_history(history, title):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'{title} - Accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.legend()
    plt.show()

# ZFNet
def build_zfnet():
    model = Sequential()
    model.add(Conv2D(64, (3,3), strides=1, activation='relu', padding='same', kernel_initializer=HeNormal(), input_shape=(img_height, img_width, 3)))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2,2), strides=2))
    model.add(Conv2D(128, (3,3), activation='relu', padding='same', kernel_initializer=HeNormal()))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2,2), strides=2))
    model.add(Conv2D(256, (3,3), activation='relu', padding='same', kernel_initializer=HeNormal()))
    model.add(BatchNormalization())
    model.add(Conv2D(256, (3,3), activation='relu', padding='same', kernel_initializer=HeNormal()))
    model.add(BatchNormalization())
    model.add(Conv2D(128, (3,3), activation='relu', padding='same', kernel_initializer=HeNormal()))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2,2), strides=2))
    model.add(Flatten())
    model.add(Dense(1024, activation='relu', kernel_initializer=HeNormal()))
    model.add(Dropout(0.5))
    model.add(Dense(512, activation='relu', kernel_initializer=HeNormal()))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    return model

# VGG16
def build_vgg16():
    base = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
    for layer in base.layers:
        layer.trainable = False
    x = Flatten()(base.output)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.5)(x)
    out = Dense(num_classes, activation='softmax')(x)
    return Model(inputs=base.input, outputs=out)

# GoogLeNet (InceptionV1)
def inception_module(x, f1, f3_in, f3_out, f5_in, f5_out, pool_proj):
    path1 = Conv2D(f1, (1, 1), padding='same', activation='relu')(x)

    path2 = Conv2D(f3_in, (1, 1), padding='same', activation='relu')(x)
    path2 = Conv2D(f3_out, (3, 3), padding='same', activation='relu')(path2)

    path3 = Conv2D(f5_in, (1, 1), padding='same', activation='relu')(x)
    path3 = Conv2D(f5_out, (5, 5), padding='same', activation='relu')(path3)

    path4 = MaxPooling2D((3, 3), strides=(1, 1), padding='same')(x)
    path4 = Conv2D(pool_proj, (1, 1), padding='same', activation='relu')(path4)

    return Concatenate()([path1, path2, path3, path4])

def build_inception():
    input_layer = Input(shape=(img_height, img_width, 3))

    x = Conv2D(64, (7, 7), strides=(2, 2), padding='same', activation='relu')(input_layer)
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    x = Conv2D(64, (1, 1), activation='relu')(x)
    x = Conv2D(192, (3, 3), padding='same', activation='relu')(x)
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    # Inception Modules
    x = inception_module(x, 64, 96, 128, 16, 32, 32)
    x = inception_module(x, 128, 128, 192, 32, 96, 64)
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    x = inception_module(x, 192, 96, 208, 16, 48, 64)
    x = inception_module(x, 160, 112, 224, 24, 64, 64)
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    x = inception_module(x, 256, 160, 320, 32, 128, 128)
    x = AveragePooling2D(pool_size=(7, 7), strides=(1, 1), padding='valid')(x)

    x = Flatten()(x)
    x = Dropout(0.4)(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.4)(x)
    output_layer = Dense(num_classes, activation='softmax')(x)

    return Model(inputs=input_layer, outputs=output_layer)


# Train and evaluate model
def run_model(model_fn, name):
    print(f"\n--- Training {name} ---")
    model = model_fn()
    model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(train_generator, validation_data=val_generator,
                        epochs=epochs, callbacks=[EarlyStopping(patience=3, restore_best_weights=True)])
    plot_history(history, name)
    loss, acc = model.evaluate(val_generator)
    print(f"{name} Accuracy: {acc * 100:.2f}% | Loss: {loss:.4f}")
    return model

# Inference timing
def evaluate_model_inference(model, data_gen, name):
    print(f"\n--- Inference Timing for {name} ---")
    total_time = 0
    num_batches = 0
    for batch_x, _ in data_gen:
        start = time.time()
        _ = model.predict(batch_x)
        end = time.time()
        total_time += (end - start)
        num_batches += 1
        if num_batches >= 10:
            break
    avg_time = total_time / num_batches
    print(f"{name} - Avg inference time per batch: {avg_time:.4f} seconds")

# Run models
zfnet_model = run_model(build_zfnet, "ZFNet")
vgg_model = run_model(build_vgg16, "VGG16")
inception_model = run_model(build_inception, "GoogLeNet (InceptionV1)")

# Compare inference times
evaluate_model_inference(zfnet_model, val_generator, "ZFNet")
evaluate_model_inference(vgg_model, val_generator, "VGG16")
evaluate_model_inference(inception_model, val_generator, "GoogLeNet (InceptionV1)")
```
